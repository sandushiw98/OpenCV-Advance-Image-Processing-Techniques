```python
from google.colab.patches import cv2_imshow
import cv2

# Assuming you have already loaded your image
image = cv2.imread('image.jpg')

if image is not None:
    # Get the dimensions of the image
    (h, w) = image.shape[:2]

    # Calculate the center of the image
    center = (w // 2, h // 2)

    # Define the rotation matrix (rotate by 45 degrees)
    rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)

    # Perform the rotation
    rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

    # Display the rotated image using cv2_imshow
    cv2_imshow(rotated_image)
else:
    print("Failed to load the image. Please check the file path.")


```

```python
from google.colab.patches import cv2_imshow
import cv2
# Crop the image (example: crop the region from (50, 50) to (200, 200))
cropped_image = image[50:200, 50:200]

# Display the cropped image
cv2_imshow(cropped_image)

```

```python
# Resize the image to 50% of its original size
resized_image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)

# Display the resized image
cv2_imshow(resized_image)

```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
cv2_imshow(gray_image)

```

```python
import cv2

# Read the image
image = cv2.imread('image.jpg', 0)  # Read in grayscale

# Apply Canny Edge Detection
edges = cv2.Canny(image, 100, 200)

# Display the result
cv2_imshow(edges)

```

```python
# Apply Gaussian Blur
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)

# Display the result
cv2_imshow(gaussian_blur)

```

```python
# Apply Median Blur
median_blur = cv2.medianBlur(image, 5)

# Display the result
cv2_imshow(median_blur)

```

```python
image = cv2.imread('image1.png')
# Create a kernel
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

# Apply Dilation
dilated_image = cv2.dilate(image, kernel, iterations=1)

# Display the result
cv2_imshow(dilated_image)

```

```python
# Apply Erosion
eroded_image = cv2.erode(image, kernel, iterations=1)

# Display the result
cv2_imshow(eroded_image)

```
