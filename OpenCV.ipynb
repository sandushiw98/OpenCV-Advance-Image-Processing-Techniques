**Reading an image and printing its dimension**

```python
import cv2
image = cv2.imread('image.jpg')
x,y = image.shape[:2]
print("Height = {}, Width = {}".format(x,y))
```

**Extracting RGB values of the pixels**

```python
(R,G,B) = image[250,250] #selecting the pixels at this location
print("R ={}, G= {}, B= {}".format(R,G,B))
B=image[250,250,0] #NumPy Arrays: In Python, images can be represented as NumPy arrays with the shape (height, width, 3),
                  #where '3' stands for the three color channels
print("Just the value of B = {}".format(B))
```

**Resizing the image to any dimension**

```python
size=cv2.resize(image, (1920,1080))
cv2.imwrite("newimage1.jpg",image)
```

**Placing a box around any pixels**

```python
rectangle = cv2.rectangle(image,(1500,900),(600,400),(255,0,0),2)
cv2.imwrite("newimage2.jpg",image)
```

**Adding text to the screen**

```python
# cv2.putText(): This is an OpenCV function used to draw text on an image.
# image: This is the image object where the text will be drawn. It should be a NumPy array that OpenCV can manipulate, typically loaded using cv2.imread() or created by other means in OpenCV.
# 'Anirudh Rao': The string of text that you want to add to the image.
# (500, 550): This tuple represents the x and y coordinates of the bottom-left corner of the text in the image. This is the starting point from where the text will be drawn.
# cv2.FONT_HERSHEY_SIMPLEX: This specifies the font type. cv2.FONT_HERSHEY_SIMPLEX is a simple, compact font. OpenCV offers several other fonts as well.
# 4: This is the font scale factor that multiplies the base font size, making the text larger or smaller.
# (255, 0, 0): This is the color of the text. In OpenCV's default BGR color format, (255, 0, 0) represents the color blue. To draw the text in red, you would use (0, 0, 255).
# 2: This is the thickness of the lines used to draw the text.
text_input = cv2.putText(image,'Anirudh Rao',(500,550),cv2.FONT_HERSHEY_SIMPLEX,4,(255,0,0),2)
cv2.imwrite("newimage3.jpg",image)
```
